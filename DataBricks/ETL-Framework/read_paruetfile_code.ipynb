```python
# Validate & correctly set storage keys for both blob and dfs endpoints
import base64, traceback

# Provide key via widget (temporary) OR via secret (recommended)
dbutils.widgets.text("use_secret", "true")   # set to "false" to use direct widget key
dbutils.widgets.text("secret_scope", "my-scope")   # change if you have a scope
dbutils.widgets.text("secret_key_name", "scrgvkrmade-key")  # change if using secret
dbutils.widgets.text("direct_key", "")  # only used if use_secret == "false"

use_secret = dbutils.widgets.get("use_secret").strip().lower() in ("true","1","yes","y")
secret_scope = dbutils.widgets.get("secret_scope").strip()
secret_key_name = dbutils.widgets.get("secret_key_name").strip()
direct_key = dbutils.widgets.get("direct_key").strip()

# The target paths you will read/write
wasbs_example_path = "wasbs://project@scrgvkrmade.blob.core.windows.net/bronze/ResellerSales/Sales.Currency/"
abfss_example_path = "abfss://project@scrgvkrmade.dfs.core.windows.net/silver/test_delta/"

# get key either from secret or direct widget
key = None
if use_secret:
    try:
        key = dbutils.secrets.get(scope=secret_scope, key=secret_key_name)
        print("Loaded key from secret scope:", secret_scope, "/", secret_key_name)
    except Exception as e:
        print("Could not load secret. Falling back to direct_key (if provided). Error:", e)
        key = direct_key or None
else:
    key = direct_key or None

if not key:
    raise Exception("No storage key provided. Provide direct_key or set use_secret=true and correct secret scope/key.")

# strip surrounding quotes/newlines and whitespace
key = key.strip()
if (key.startswith('"') and key.endswith('"')) or (key.startswith("'") and key.endswith("'")):
    key = key[1:-1]
key = key.strip()

print("Key length:", len(key))

# quick base64 decode check (storage keys are base64-like)
valid_base64 = True
try:
    decoded = base64.b64decode(key)
    if len(decoded) < 16:
        print("Warning: decoded key length looks short:", len(decoded))
except Exception as e:
    valid_base64 = False
    print("Warning: key failed base64 decode check:", e)

print("Base64-valid:", valid_base64)

# Now set both spark.conf keys (blob + dfs)
account = "scrgvkrmade"
try:
    # blob
    blob_conf = f"fs.azure.account.key.{account}.blob.core.windows.net"
    spark.conf.set(blob_conf, key)
    print("Set spark.conf:", blob_conf)
    # dfs
    dfs_conf = f"fs.azure.account.key.{account}.dfs.core.windows.net"
    spark.conf.set(dfs_conf, key)
    print("Set spark.conf:", dfs_conf)
except Exception:
    traceback.print_exc()
    raise RuntimeError("Failed to set spark.conf for storage account keys. Check key value and cluster permissions.")

# Quick checks: list small folder (read) and write small delta (write test)
print("\nAttempting read test (list 1 file from WASBS path)...")
try:
    files = dbutils.fs.ls(wasbs_example_path)
    print("Read/list succeeded. Example entries:")
    for f in files[:5]:
        print(" -", f.path)
except Exception as e:
    print("Read/list failed:", e)

print("\nAttempting small delta write test to ABFSS path (will overwrite if exists)...")
try:
    # small DF and write to test location
    test_df = spark.range(1).withColumnRenamed("id", "test_col")
    test_df.write.format("delta").mode("overwrite").save(abfss_example_path)
    print("Small delta write succeeded.")
    # try to read back
    read_back = spark.read.format("delta").load(abfss_example_path)
    print("Read back delta rows:", read_back.count())
    display(read_back)
except Exception as e:
    print("Small delta write/read test FAILED:", e)
    traceback.print_exc()

print("\nIf the write failed with 'Invalid configuration value' then the key is malformed or not the correct storage-account key.")
print("If the read failed, ensure the key matches the blob endpoint and account name.")
print("\nRecommended: use the SAME storage key from Azure Portal -> Storage Account -> Access keys (Key1 or Key2).")
```
